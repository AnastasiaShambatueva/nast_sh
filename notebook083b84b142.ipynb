**АНАЛИТИКА:**

```python
df['Duration'] = df['Duration_ms'].apply(lambda x: round(x/(1000*60))).drop(columns = ['Duration_ms'])
df.head()
```

```python
# Во-первых, стоит определить корреляцию между всеми характеристиками трека.
# Для этого построим тепловую карту корреляции.
corr_df = df.loc[:, ['Popularity', 'Duration', 'Explicit', 'Danceability', 'Energy', 'Key', 'Loudness',
'Mode', 'Speechiness', 'Acousticness', 'Instrumentalness', 'Liveness',
'Valence', 'Tempo', 'Time_signature']]
corr_df_1 = corr_df.corr()
fig, ax = plt.subplots(figsize=(12, 8))
sns.heatmap(data=corr_df_1, fmt='.2f', annot=True, ax=ax)
plt.tight_layout()
plt.show()
```

**Вывод:**

На данной карте больше всего коррелируют между собой следующие характеристики:
* Громкость и энергичность - энегричность трека зависит от его громкости, чем громче трек, тем более энергичен он будет;
* Позитивность и танцевальность - трек лучше подходит для танцев, если обладает позитивным настроем.

Также коррелируют между собой количество текста в треке и наличие в нем откровенных текстов. Можно сказать, что чем больше текста в треке, тем больше будет нецензурных выражений.
В остальных случаях корреляция наблюдается, но уже очень слабая.

```python
# Так как на корреляционной карте была отражена зависимость между энергичностью и громкостью трека,
# покажем эту зависимоть подробнее на следующем графике.
plt.scatter(x=df['Loudness'], y=df['Energy'], alpha=0.1, color='#C71585') 
plt.title('Зависимость энергичности трека от его громкости')
plt.xlabel('Громкость')
plt.ylabel('Энергичность')
pass
```

```python
# Создадим простую модель, содержащую 10 эпох и посмотрим на результаты обучения.
import numpy as np
from sklearn.model_selection import train_test_split 
from sklearn.datasets import make_classification 
from sklearn.preprocessing import StandardScaler 
from keras.models import Sequential
from keras.layers import Dense


x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

model = Sequential()
model.add(Dense (32, activation='relu', input_shape=(x_train.shape[1],)))
model.add(Dense (16, activation='relu'))
model.add(Dense (1, activation='sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.fit(x_train, y_train, epochs=10, batch_size=32, validation_split=0.2)

loss, accuracy = model.evaluate(x_test, y_test)
print(f'Test Accuracy: {accuracy:.2f}')
```
